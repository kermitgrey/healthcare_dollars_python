<h3>Financial Relationships in Health Care</h3>

<p> Early in the 2000â€™s, the United States Congress became increasingly concerned about the rise in the financial relationships between pharmaceutical companies and physicians and its possible adverse impact on patient care.  As a consequence, Congress passed the Physician Payments Sunshine Act in 2010 to promote the transparency of any financial relationships between health care providers and pharmaceutical companies.  The act gave the Centers for Medicare and Medicaid Services (CMS) statutory responsibility for collecting and tracking any transfers-of-value between companies and physicians.</p>

<p> CMS has developed a collection and tracking system called the Open Payments System to carry out its statutory responsibility.  Within that system, companies submit their payment transactions using a set of reporting attributes that serve to describe any transfers-of-value made to physicians.  The three major categories of data collected in Open Payments are:</p>

<ul>
    <li>General Payments: payments not connected to a research agreement.</li>
    <li>Research Payments: payments made in connection with a research agreement.</li>
    <li>Physician Ownership Information: ownership/investment interests physicians or their immediate family have in a reporting company.</li>
</ul>

<p> The presentation below is a descriptive analysis of the General Payments data 
    only and covers the years 2016-2019.  Moreover, it restricts itself to payments 
    pertaining only to physicians residing in the United States (the 50 states, 
    District of Columbia, and Puerto Rico).</p>

<h4> Conclusion</h4>

<p> The Open Payments System shows the large volume of activity comprising
    the financial flows between companies and physicians. However, as 
    demonstrated above, the greatest proportion of that flow is concentrated
    in a relatively small number of participants.</p>
    
<p> When you segment physicians by total receipts, you discern that 81 percent 
    of those physicians receive less than \$1,000 per year and another
    15 percent receive between \$1,000 and \$10,000.  Only a small proportion
    of the physicians fall into the highly compensated segment.  Another 
    characteristic of the population of physicians is that the largest
    proportion of compensation is directed to physicians belonging to three 
    specialties, Internal Medicine, Orhtopaedic Surgery, and Psychiatry & Neurology. </p>
    
<p> From the company perspective, the distribution of spending is also skewed.
    As you saw, around 84 percent of the companies spend less than \$1M per year while
    the vast minority spend an amount in excess of \$1M.  But, it is this minority
    that comprises the vast majority of total spending.  Moreover, 69 percent of
    that spending is directed in three areas, Consulting, Services, and Royalty.</p>
    
<p> The general conclusion you can make is that companies are not spreading their
    dollars evenly throughout the health care market. Instead, they concentrate their 
    dollars in select markets and direct there largest dollar amounts to a small 
    subset of the physician population.  And, what they are purchasing is a
    direct service in the case of Consulting and Services or a form of
    intellectual property via Royalty.</p>
